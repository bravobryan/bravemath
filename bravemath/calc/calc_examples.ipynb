```python
# Example of annotating angles in a triangle using matplotlib
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc

fig, ax = plt.subplots()

# vertices
vertices = np.array([[0, 0], [2, 1], [1, 0]])

polygon = plt.Polygon(vertices, closed=True, fill=None, edgecolor='g')
ax.add_patch(polygon)
ax.text(0.5, -0.05, 'a', fontsize=10, color='b')
ax.text(1.5, 0.4, 'b', fontsize=10, color='b')
ax.text(1, 0.6, 'c', fontsize=10, color='b')

plt.plot([1, 2], [0, 0], 'r--')

# Annotate theta angle with an arc
arc = Arc((1, 0), 0.5, 0.5, angle=0, theta1=45, theta2=180, color='g', linestyle='--')
ax.add_patch(arc)
ax.text(0.85, 0.03, r'$\theta$', fontsize=15, color='g')

arc = Arc((1, 0), 0.75, 0.75, angle=0, theta1=0, theta2=45, color='r', linestyle='--')
ax.add_patch(arc)
ax.text(1.1, .03, r'$\phi$', fontsize=15, color='r')


ax.set_xlim(-.05, 2.05)
ax.set_ylim(-0.1, 1.1)
ax.set_aspect('equal', adjustable='box')

plt.show()
```
