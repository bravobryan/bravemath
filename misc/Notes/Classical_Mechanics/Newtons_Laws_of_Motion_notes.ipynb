### Space
A way to express a point in space `P` in a 3-D space is to use a position vector `r` from a chosen origin `O`.
- The origin `O` is expressed as 
- A way to express `r` is by introducing three unit vectors `x_hat`, `y_hat`, and `z_hat`.
    - The notation of `r` is usually written as `r = (x*x_hat) + (y*y_hat) + (z*z_hat)`.
    - 
- It is also convenient to abbreviate the notation as `r = (x, y, z)`

```python
# Graph a point in 3D space using matplotlib
import matplotlib.pyplot as plt
# from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
point = np.array([1, 1, 1])  # Setting point P at (1, 1, 1)
ax.scatter(point[0], point[1], point[2], color='b', s=100)
# Add arrow that from the origin of the 3D space to the point
ax.quiver(0, 0, 0, point[0], point[1], point[2], color='r', arrow_length_ratio=0.1)
# Extend the axes to better visualize the point
ax.set_xlim([0, 1.1])
ax.set_ylim([0, 1.1])
ax.set_zlim([0, 1.1])
# Annotate the Origin of the 3D space as 'O' and the point as 'P'
ax.text(0, 0.1, 0, 'Origin `Oxyz`', color='k', fontsize=12)
ax.text(point[0], point[1], point[2]+0.1, 'Point `P`', color='k', fontsize=12)
# Annotate the vector as 'r'
ax.text(point[0]/2, 0, point[2]/2, 'Vector `r`', color='r', fontsize=12)

# Rotate the view for better visualization
ax.view_init(elev=20., azim=30)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
# Add title
ax.set_title('Position Vector `r = (x, y, z)`', fontsize=14)
plt.show()
```
